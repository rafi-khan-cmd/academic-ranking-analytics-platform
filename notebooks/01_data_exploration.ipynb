# Data Exploration

This notebook explores the raw data from OpenAlex API, examining institution coverage, data quality, and initial patterns.

```python
import pandas as pd
import numpy as np
import json
from pathlib import Path
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)
```

```python
# Load raw institutions data
data_dir = Path('../data/raw')

with open(data_dir / 'institutions_raw.json', 'r') as f:
    institutions = json.load(f)

print(f"Total institutions: {len(institutions)}")
```

```python
# Convert to DataFrame
df = pd.json_normalize(institutions)
df.head()
```

```python
# Explore country distribution
if 'country_code' in df.columns:
    country_counts = df['country_code'].value_counts().head(20)
    plt.figure(figsize=(12, 6))
    country_counts.plot(kind='bar')
    plt.title('Top 20 Countries by Institution Count')
    plt.xlabel('Country Code')
    plt.ylabel('Number of Institutions')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```
